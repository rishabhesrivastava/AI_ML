Y = theta0 + theta1 . X

```python
# Importing library and loading the file column in varibale X and Y
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

# Load the CSV
data = pd.read_csv('Salary_Data.csv')
X = data.iloc[:, 0].values  # Feature (YearsExperience)
Y = data.iloc[:, 1].values  # Target (Salary)
```

```python
# Splitting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
m = len(X_train)        # Number of samples
print(m)
```

```python
# Declaring variable and setting up initial values. 

# y = theta0 + theta1 * x

theta0 = 0.0      # Intercept Initial Value
theta1 = 0.0      # Slope Initial Value
alpha = 0.01      # Learning rate
epochs = 1000     # Number of iterations
```

```python
for i in range(epochs):
    Y_pred = theta1 * X_train + theta0
    # Compute gradients
    d_theta1 = (-2/m) * np.sum(X_train * (y_train - Y_pred))
    d_theta0 = (-2/m) * np.sum(y_train - Y_pred)
    # Update parameters
    theta1 = theta1 - alpha * d_theta1
    theta0 = theta0 - alpha * d_theta0
    
print(f"Slope (theta1): {theta1}")
print(f"Intercept (theta0): {theta0}")

Y_pred_test = theta1 * X_test + theta0
```

```python
# 6. Plot results
plt.scatter(X_train, y_train, color='blue', label='Training Data')
plt.scatter(X_test, y_test, color='green', label='Test Data')
plt.scatter(X_test, Y_pred_test, color='orange', marker='x', label='Predicted Test Y')
plt.plot(X, theta1 * X + theta0, color='red', label='Regression Line')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.title('Linear Regression using Gradient Descent')
plt.legend()
plt.show()
```
